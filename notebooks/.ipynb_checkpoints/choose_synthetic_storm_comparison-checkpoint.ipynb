```python
%config InlineBackend.figure_format = 'retina'
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import numpy as np
import pygmt
import pandas as pd
import xarray as xr
import datetime
```

```python
# Import NACCS_TS_Sim0_Post0_ST_Trop_Param.csv to get landfall location for each storm
# Storms 480-490
storm_param_df = pd.read_csv('/home/catherinej/NACCS_storm_input_files/NACCS_TS_Sim0_Post0_ST_TROP_Param.csv')
storm_param_df = storm_param_df.iloc[2:]
storm_param_df['Storm ID'] = storm_param_df['Storm ID'].astype('int')
```

```python
# Subset the storm_param_df to get only the 11 storms
storms_subset = storm_param_df[storm_param_df['Storm ID'].between(480,490)]
storms_subset = storms_subset.drop(['Unnamed: 9'], axis=1)
storms_subset
```

```python
# Loop through all storm_files and get the landfall time to 
# match reference lat/lon and append to storms_subset to get landfall time in seconds for each storm
landfall_time = []
for storm_no in storms_subset['Storm ID']:
    storm_file = f'NACCS_TP_{storm_no:04}_SYN_L2.nc'
    print('Storm file: ', storm_file)
    ds = xr.open_dataset(os.path.join('/home/catherinej/storm_files', storm_file))
    eye = ds.eye_loc.data
    lat_row = np.where(eye[:,1] >= 40.68)
    lon_row = np.where(eye[:,0] <= -73.31)
    idx = max(min(lat_row[0]), min(lon_row[0]))
    print(ds.time[idx].values)
    landfall_time.append(ds.time[idx].values)
storms_subset['Landfall Time (sec)'] = landfall_time
storms_subset.reset_index()
```

```python
# Find all files for NOAA Gauge 8531680 ** shortened in geoclaw to 85310 due to number constraints
import glob
PATH = '/home/catherinej/BarrierBreach/data/synthetic_sim_sandy_hook_gauges/'
filename = '*.txt'
allfiles = glob.glob(os.path.join(PATH,'**',filename), recursive=True)
allfiles = [x for x in allfiles if 'radius_trials' not in x]
col_names = ['time', 'eta']
allfiles
```

```python
# Load all gauge files into a single dataframe
allfiles
df_dict = {}
storm_nos = []
for file in allfiles:
    fldr = file.split('/')[-1].split('_')[0]
    storm_nos.append(fldr)
    cols = [f'{fldr}_{i}' for i in col_names]
    df = pd.read_csv(file, skiprows=4, header=None, 
                               delim_whitespace=True, usecols=[1,5], names=col_names)
    # df['storm_no'] = int(fldr)
    df_dict[fldr] = df
# gauge_df = pd.concat(df_list, axis=1, keys=storm_nos)
```

```python
df_dict['482']
```

```python
# Update time column to be in hours before/after landfall time
for key in df_dict:
    # gauge_df.xs(str(num), level=0, axis=1)
    landfall_time = storms_subset.loc[storms_subset['Storm ID'] == int(key),
                                      'Landfall Time (sec)'].iloc[0]

    df_dict[key]['time'] -= landfall_time
    df_dict[key]['time'] /= 3600.0
# gauge_df.loc[:,(slice(None), 'time')] = gauge_df.loc[:,slice(None, 'time')]/3600
```
