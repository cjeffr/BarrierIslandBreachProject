```python
import pandas as pd

# Load the data from the pickle.gz file
df = pd.read_pickle('~/BarrierBreach/data/breach_data_updated.pkl.gz')

# Group the data by 'key' column
grouped_data = df.groupby('key')

# Initialize a list to store the extracted data
summary_data = []

# Helper function to extract simulation type from the 'key' column
def extract_simulation_type(key):
    key_lower = key.lower()
    if 'w_' in key_lower and 'd' not in key_lower:
        return 'Width'
    elif 'd' in key_lower and 'w' not in key_lower:
        return 'Depth'
    elif 'loc' in key_lower:
        return 'Location'
    elif 'wd' in key_lower or 'dw' in key_lower:
        return 'Width & Depth'
    elif 'west' in key_lower:
        return 'West'
    elif 'east' in key_lower:
        return 'East'
    else:
        return 'Unknown'

# Calculate statistics for 'Width' simulation type
width_data = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'Width']
num_width_breaches = len(width_data)
min_width = width_data['breach_width (m)'].min()
max_width = width_data['breach_width (m)'].max()
mean_width = width_data['breach_width (m)'].mean()
median_width = width_data['breach_width (m)'].median()

# Calculate statistics for 'Depth' simulation type
depth_data = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'Depth']
num_depth_breaches = len(depth_data)
min_depth = depth_data['Depth'].min()
max_depth = depth_data['Depth'].max()
mean_depth = depth_data['Depth'].mean()
median_depth = depth_data['Depth'].median()

# Calculate statistics for 'Location' simulation type
loc_data = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'Location']
num_loc_breaches = len(loc_data)
min_loc_width = loc_data['breach_width (m)'].min()
max_loc_width = loc_data['breach_width (m)'].max()
mean_loc_width = loc_data['breach_width (m)'].mean()
median_loc_width = loc_data['breach_width (m)'].median()

# Calculate statistics for 'Location' simulation type
east_data = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'East']
num_east_breaches = len(east_data)
min_east_width = east_data['breach_width (m)'].min()
max_east_width = east_data['breach_width (m)'].max()
mean_east_width = east_data['breach_width (m)'].mean()
median_east_width = east_data['breach_width (m)'].median()

# Calculate statistics for 'Location' simulation type
west_data = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'West']
num_west_breaches = len(west_data)
min_west_width = west_data['breach_width (m)'].min()
max_west_width = west_data['breach_width (m)'].max()
mean_west_width = west_data['breach_width (m)'].mean()
median_west_width = west_data['breach_width (m)'].median()

# Iterate through the grouped data for other simulation types
for key, group in grouped_data:
    try:
        # Calculate statistics for each individual simulation
        num_breaches = len(group)
        min_width = group['breach_width (m)'].min()
        max_width = group['breach_width (m)'].max()
        mean_width = group['breach_width (m)'].mean()
        median_width = group['breach_width (m)'].median()

        min_depth = group['Depth'].min()
        max_depth = group['Depth'].max()
        mean_depth = group['Depth'].mean()
        median_depth = group['Depth'].median()

        # Calculate statistics for the type of simulation (width, depth, location, etc.)
        sim_type = extract_simulation_type(key)

        # Append the extracted data to the summary_data list
        summary_data.append({
            'Key': key,
            'Simulation Type': sim_type,
            'Total Breaches': num_breaches,
            'Min Width': min_width,
            'Max Width': max_width,
            'Mean Width': mean_width,
            'Median Width': median_width,
            'Min Depth': min_depth,
            'Max Depth': max_depth,
            'Mean Depth': mean_depth,
            'Median Depth': median_depth
        })
    except KeyError as e:
        print(f"KeyError occurred for key: {key}")
        print(e)
        continue  # Skip this key and proceed with the next one

# Convert the summary_data list to a new DataFrame and display it
summary_df = pd.DataFrame(summary_data)
print(summary_df.head())
```

```python
import pandas as pd

# Load the data from the pickle.gz file
df = pd.read_pickle('~/BarrierBreach/data/breach_data_updated.pkl.gz')

# Group the data by 'key' column
grouped_data = df.groupby('key')

# Helper function to extract simulation type from the 'key' column
# Helper function to extract simulation type from the 'key' column
def extract_simulation_type(key):
    key_lower = key.lower()
    if 'w_' in key_lower and 'd' not in key_lower:
        return 'Width'
    elif 'd' in key_lower and 'w' not in key_lower:
        return 'Depth'
    elif 'loc' in key_lower:
        return 'Location'
    elif 'wd' in key_lower or 'dw' in key_lower:
        return 'Width & Depth'
    elif 'west' in key_lower:
        return 'West'
    elif 'east' in key_lower:
        return 'East'
    else:
        return 'Unknown'


# Create separate DataFrames for each simulation type
width_df = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'Width']
depth_df = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'Depth']
location_df = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'Location']
width_depth_df = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'Width & Depth']
east_df = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'East']
west_df = df[df['key'].apply(lambda key: extract_simulation_type(key)) == 'West']
# Define a function to format numeric values as LaTeX-friendly strings
def format_latex_number(value):
    if isinstance(value, (int, float)):
        return f"{value:.2f}"
    else:
        return str(value)

# Helper function to generate LaTeX table from DataFrame
def df_to_latex(df):
    # Use df.describe() to get summary statistics and transpose the result
    summary = df.describe().T

    # Format numeric values as LaTeX-friendly strings
    summary = summary.applymap(format_latex_number)

    # Add a 'Simulation Type' column to the DataFrame
    summary['Simulation Type'] = df['Simulation Type'].iloc[0]

    # Reorder columns for desired output
    summary = summary[['Simulation Type', 'count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]

    # Reset the index to get the default integer index
    summary.reset_index(drop=True, inplace=True)

    return summary

# Convert each DataFrame to a LaTeX table
latex_width_table = df_to_latex(width_df)
latex_depth_table = df_to_latex(depth_df)
latex_location_table = df_to_latex(location_df)
latex_width_depth_table = df_to_latex(width_depth_df)
latex_east_table = df_to_latex(east_df)
latex_west_table = df_to_latex(west_df)

# Concatenate the separate DataFrames vertically
all_summary_df = pd.concat([latex_width_table, latex_depth_table, 
                            latex_location_table, latex_width_depth_table,
                           latex_east_table, latex_west_table], ignore_index=True)

# Save the concatenated DataFrame to a LaTeX table file
with open('summary_table.tex', 'w') as f:
    f.write(all_summary_df.to_latex(index=False, escape=False))
```
