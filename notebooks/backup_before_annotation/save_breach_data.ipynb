```python
import pandas as pd
import glob
import os
import numpy as np
```

```python
def calc_width(lat, lon1, lon2):
    m_per_deg = np.pi/180 * 6371000* np.cos(lat * np.pi/180)
    dist = lon2 - lon1
    width = dist * m_per_deg
    return width
```

```python

def find_folders_with_breach_data(root_dir):
    breach_data_files = []
    for root, dirs, files in os.walk(root_dir):
        for folder_name in dirs:
            # Check if the folder name matches any of the desired names
            if folder_name in simulation_names:
                print(folder_name)
                folder_path = os.path.join(root, folder_name)
                # Search for the 'breach.data' file inside the current folder
                breach_data_file_path = glob.glob(os.path.join(folder_path, 'breach.data'))
                if breach_data_file_path:
                    # Append the path to the 'breach.data' file to the list
                    breach_data_files.append(breach_data_file_path[0])
    return breach_data_files

PATH = '/projects/weiszr_lab/catherine/breach_sims'
# simulation_names = ['folder1', 'folder2', 'folder3']

breach_data_files = find_folders_with_breach_data(PATH)
print(breach_data_files)
```
